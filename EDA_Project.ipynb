# EDA PROJECT ON INDIAN WEATHER REPOSITORY

India's weather is renowned for its complexity, influenced by diverse geographical features, varying landscapes, and climatic zones. This EDA venture seeks to dissect this complexity, leveraging a comprehensive dataset spanning several years, encompassing multiple regions, and capturing an array of meteorological parameters. In this project, I've  used dataset taken from Kaggle. The link for the dataset is as follows : https://www.kaggle.com/datasets/nelgiriyewithana/indian-weather-repository-daily-snapshot

Exploratory Data Analysis, or EDA, is an important step in any Data Analysis or Data Science project. EDA is the process of investigating the dataset to discover patterns, and anomalies (outliers), and form hypotheses based on our understanding of the dataset.

#### 4. MISSING VALUE IMPUTATION

##### We will now check for missing values in our dataset. In case there are any missing entries, we will impute them with appropriate values (mode in case of categorical feature, and median or mean in case of numerical feature). We will use the isnull() function for this purpose

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['country','latitude','longitude','last_updated_epoch','last_updated','condition_text','wind_degree'], axis=1)
df.head(5)
```

```python
duplicate_rows_df = df[df.duplicated()]
duplicate_rows_df 
```

```python
df.count()  
```

```python
df = df.drop_duplicates()
df.head()
```

### We can do some sorting operations

```python
df = df.sort_values(by='temperature_celsius', ascending=False)
highest_temp_row = df.iloc[0]
location_name = highest_temp_row['location_name']
region = highest_temp_row['region']
timezone = highest_temp_row['timezone']
temp_celsius = highest_temp_row['temperature_celsius']
print(f"The city with the highest temperature in celsius is {location_name}, {region}  with a temperature of {temp_celsius}°C. ")
```

##### Correlation between temperature_celcius & humidity

```python
corr = df['temperature_celsius'].corr(df['humidity'])
print('The correlation between temperature_celsius and humidity is',corr)
```

##### Locations where the 'air_quality_Carbon_Monoxide' is above a certain threshold

```python
a=df['air_quality_Carbon_Monoxide'].mean()
a
```

```python
threshold =a
high = df[df['air_quality_Carbon_Monoxide']> threshold]
if not high.empty:
    print("Location with 'air_quality_Carbon_monoxide'above the threshold: ")
    print(high[['location_name','air_quality_Carbon_Monoxide']])
else:
    print("No Location Have 'air_quality_Carbon_Monoxide' above  the threshold ")
```

##### The average  'temperature_celsius' across all locations

```python
average = df['temperature_celsius'].mean()
print(f'The average temperature in celsius across all location is {average:.2f} degree.')
```

##### DataFrame where 'temperature_celsius' and 'air_quality_Carbon_Monoxide' are greater than mean

```python
filtered_values = np.where((df['temperature_celsius']>=average) & (df['air_quality_Carbon_Monoxide']> a) )
print(filtered_values)
# display(df.loc[filtered_values])
```

#### 5. GRAPHICAL REPRESENTATION

```python
sns.boxplot(df['humidity'],color='yellow')
```

```python
top_20 = df['region'].value_counts()[:20]
top_20.plot(kind='bar',figsize=(10,8))
plt.title('All top 20 regions')
```

```python
x = df['temperature_celsius']
y = df['humidity']
plt.figure(figsize=(8, 6))
plt.scatter(x, y,c='green', alpha=0.5)
plt.title("Scatter Plot of Temperature (Celsius) vs. Humidity")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Humidity")
plt.grid(True)
plt.show()
```

##### The correlation coefficient of 0.5 suggest that a very weak or negligible correlation between temperature_celsius and humidity

```python
location = df['location_name'].head(10)
temps_celsius = df['temperature_celsius'].head(10)
```

```python
plt.plot(temps_celsius,location,c='purple')
plt.title('Temperature in celsius for Top 10 city ')
plt.ylabel('City')
plt.xlabel('Temperature (°C)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
### Distribution of 'air_quality_Carbon_Monoxide' in High CO Locations
high = df[df['air_quality_Carbon_Monoxide'] > threshold]
plt.hist(high['air_quality_Carbon_Monoxide'], bins=20,color='lightpink')
plt.title("Distribution of 'air_quality_Carbon_Monoxide' in High CO Locations")
plt.xlabel("Carbon Monoxide Levels")
plt.ylabel("Frequency")
plt.show()
```
